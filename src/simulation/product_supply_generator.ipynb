```python
import random
import pandas as pd
```

```python
class ProductSupplierMapper:

    def assign_suppliers(
        self,
        products_df: pd.DataFrame,
        suppliers_df: pd.DataFrame
    ) -> pd.DataFrame:

        supplier_ids = suppliers_df["supplier_id"].tolist()

        products_df = products_df.copy()

        # Assign a random supplier to each product
        # Using apply to ensure all products get a supplier assignment
        products_df["supplier_id"] = [
            random.choice(supplier_ids)
            for _ in range(len(products_df))
        ]

        return products_df
```

```python
if __name__ == "__main__":

    products = pd.DataFrame({
        "product_id": [1, 2, 3],
        "supplier_id": [None, None, None]
    })

    suppliers = pd.DataFrame({
        "supplier_id": [101, 102, 103]
    })

    mapper = ProductSupplierMapper()

    result = mapper.assign_suppliers(
        products,
        suppliers
    )

    print(result)
```
