```python
import random
from datetime import datetime, timedelta

import pandas as pd

from config import SimulationConfig
```

```python
class DisruptionGenerator:

    DISRUPTION_TYPES = [
        "PORT_DELAY",
        "LABOR_STRIKE",
        "TRANSPORT_ISSUE",
        "RAW_MATERIAL_SHORTAGE"
    ]

    def __init__(self):
        self.config = SimulationConfig()

    def generate(
        self,
        suppliers_df: pd.DataFrame
    ) -> pd.DataFrame:

        disruptions = []

        total_suppliers = len(suppliers_df)

        disruption_count = max(
            1,
            int(
                total_suppliers *
                self.config.DISRUPTION_RATE
            )
        )

        impacted_suppliers = suppliers_df.sample(
            n=disruption_count,
            replace=False
        )

        for _, supplier in impacted_suppliers.iterrows():

            start_date = datetime.now().date()

            duration_days = random.randint(
                3,
                14
            )

            end_date = (
                start_date +
                timedelta(days=duration_days)
            )

            disruptions.append({
                "supplier_id":
                    supplier["supplier_id"],

                "disruption_type":
                    random.choice(
                        self.DISRUPTION_TYPES
                    ),

                "disruption_start_date":
                    start_date,

                "disruption_end_date":
                    end_date,

                "severity_score":
                    round(
                        random.uniform(
                            0.2,
                            1.0
                        ),
                        2
                    ),

                "delay_days":
                    random.randint(
                        1,
                        10
                    ),

                "disruption_status":
                    "ACTIVE",

                "created_at":
                    datetime.now()
            })

        return pd.DataFrame(disruptions)
```

```python
if __name__ == "__main__":

    suppliers = pd.DataFrame({
        "supplier_id": range(1, 41)
    })

    generator = DisruptionGenerator()

    df = generator.generate(
        suppliers
    )

    print(df)
```
