```python
from datetime import datetime

import pandas as pd

from config import (
    SimulationConfig,
    DCS
)
```

```python
class DistributionCenterGenerator:

    def __init__(self):
        self.config = SimulationConfig()

    def generate(self) -> pd.DataFrame:
        dcs = []

        for idx, (dc_code, dc_name) in enumerate(
            DCS,
            start=1
        ):

            dc = {
                "dc_code": dc_code,
                "dc_name": dc_name,
                "state_id": f"REGION_{idx}",
                "capacity_units":
                    self.config.DC_CAPACITY_UNITS,
                "effective_start_date": datetime.now(),
                "effective_end_date": None,
                "is_current": True,
                "created_at": datetime.now(),
                "updated_at": datetime.now(),
                "change_type": "INSERT",
                "is_deleted": False
            }

            dcs.append(dc)

        return pd.DataFrame(dcs)


```

```python
if __name__ == "__main__":
    generator = DistributionCenterGenerator()

    df = generator.generate()

    print(df)

    print(
        f"\nGenerated {len(df)} distribution centers"
    )
```
