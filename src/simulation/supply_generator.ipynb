```python
import random
from datetime import datetime

import pandas as pd

from config import (
    SimulationConfig,
    SUPPLIER_CATEGORIES
)
```

```python
class SupplierGenerator:

    def __init__(self):
        self.config = SimulationConfig()

    def generate(self) -> pd.DataFrame:
        suppliers = []

        for supplier_num in range(
            1,
            self.config.NUM_SUPPLIERS + 1
        ):

            supplier = {
                "supplier_id": supplier_num,
                "supplier_code": f"SUP{supplier_num:03d}",
                "supplier_name": f"Supplier {supplier_num}",
                "supplier_category": random.choice(
                    SUPPLIER_CATEGORIES
                ),
                "lead_time_days": random.randint(
                    self.config.MIN_LEAD_TIME_DAYS,
                    self.config.MAX_LEAD_TIME_DAYS
                ),
                "risk_score": round(
                    random.uniform(
                        self.config.MIN_RISK_SCORE,
                        self.config.MAX_RISK_SCORE
                    ),
                    2
                ),
                "supplier_status": "ACTIVE",
                "effective_start_date": datetime.now(),
                "effective_end_date": None,
                "is_current": True,
                "created_at": datetime.now(),
                "updated_at": datetime.now(),
                "change_type": "INSERT",
                "is_deleted": False
            }

            suppliers.append(supplier)

        return pd.DataFrame(suppliers)
```

```python
if __name__ == "__main__":
    generator = SupplierGenerator()

    df = generator.generate()

    print(df.head())

    print(
        f"\nGenerated {len(df)} suppliers"
    )
```
