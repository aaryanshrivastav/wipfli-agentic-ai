```python
from pyspark.sql.functions import current_timestamp, lit
import uuid
batch_id = str(uuid.uuid4())
df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv("/Volumes/workspace/ecommerce/wipfli_task_1/calendar.csv")
)
df = (
    df.withColumn("ingestion_timestamp", current_timestamp())
      .withColumn("source_file", lit("calendar.csv"))
      .withColumn("source_system", lit("M5"))
      .withColumn("load_batch_id", lit(batch_id))
)
(
    df.write
      .format("delta")
      .mode("append")
      .saveAsTable("agentdb.bronze.calendar_raw")
)
```
