```python
from pyspark.sql.functions import (
    current_timestamp,
    monotonically_increasing_id
)

calendar_df = spark.table(
    "agentdb.silver.calendar"
)

dim_calendar = (
    calendar_df
    .drop("created_at")
)

(
    dim_calendar.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(
        "agentdb.gold.dim_calendar"
    )
)

print(
    f"Calendar Rows: {dim_calendar.count():,}"
)
```
